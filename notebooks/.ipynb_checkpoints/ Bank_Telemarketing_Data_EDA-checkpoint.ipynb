##### Statistic Significance

```python
# Check statistic significance
# Start by running chi2 test on the matrix and then post hoc
# Print the p-values and corrected p-values (after FDR correction)
from itertools import combinations
from scipy.stats import chi2_contingency
from statsmodels.sandbox.stats.multicomp import multipletests

def get_asterisks_for_pval(p_val):
    """Receives the p-value and returns asterisks string."""
    if p_val > 0.05:
        p_text = "ns"  # above threshold => not significant
    elif p_val < 1e-4:  
        p_text = '****'
    elif p_val < 1e-3:
        p_text = '***'
    elif p_val < 1e-2:
        p_text = '**'
    else:
        p_text = '*'
    
    return p_text

def chisq_and_posthoc_corrected(df):
    chi2, p, dof, ex = chi2_contingency(df, correction=True)
    print(f"Chi2 result of the contingency table: {chi2}, p-value: {p}")
    all_combinations = list(combinations(df.index, 2))  # gathering all combinations for post-hoc chi2
    p_vals = []
    print("Significance results:")
    for comb in all_combinations:
        new_df = df[(df.index == comb[0]) | (df.index == comb[1])]
        chi2, p, dof, ex = chi2_contingency(new_df, correction=True)
        p_vals.append(p)
        # print(f"For {comb}: {p}")  # uncorrected

    # checking significance
    # correction for multiple testing
    reject_list, corrected_p_vals = multipletests(p_vals, method='fdr_bh')[:2]
    for p_val, corr_p_val, reject, comb in zip(p_vals, corrected_p_vals, reject_list, all_combinations):
        print(f"{comb}: p_value: {p_val:5f}; corrected: {corr_p_val:5f} ({get_asterisks_for_pval(p_val)}) reject: {reject}")
```

__Hypothesis__:
  - H0 : Subscription and balance groups have no relationship (independent).
  - H1 : There is a relationship (dependent) between Subscription and balance groups.

```python
balance_group_data = pd.crosstab(bank_cleaned.balance_group, bank_cleaned.subscription)
balance_group_data
```

```python
chisq_and_posthoc_corrected(balance_group_data)
```

The difference between the subscription rate for the high balance group and the rest of the groups was found to be statistically significant with a p-value << .001. 

#### % Subscribed By This Campaign Contact Numbers <a id='%Subscribed_By_This_Campaign_Contact_Numbers'></a>

```python
#Use the data to prepare for the bins

bank_cleaned['campaign'].describe()
```

```python
#Since the data is highly imbalanced, putting it into bins is better for us to see the trend
bins = [1, 2, 3, 4, 6, 100]
category = ['1', '2', '3','4-6', '>6']

bank_cleaned['campaign_contact_number'] = pd.cut(bank_cleaned['campaign'], bins, labels = category)
contact_counts = bank_cleaned['campaign_contact_number'].value_counts().rename_axis('group').to_frame('counts')
contact_counts = contact_counts.sort_values(by = 'group', ascending = True)
contact_counts
```

```python
#Calculate the subscription rate within each group

within_campaign_group = pd.crosstab(bank_cleaned.campaign_contact_number,bank_cleaned.subscription).apply(lambda row: row/row.sum(), axis=1)
within_campaign_group.columns = ['within_no','within_yes']
within_campaign_group
```

```python
#plot the dataframe
fig, ax = plt.subplots()
fig.set_size_inches(9,6)
fontsize = 15

ax.bar(within_campaign_group.index, within_campaign_group['within_yes'],
                    alpha=0.6)

            
ax.set_ylabel('% subscribed')
ax.set_xlabel('Number of contacts')

        
ax2 = plt.twinx()
ax2.plot(contact_counts.index, contact_counts, marker='*', color= 'green',alpha=0.6)
#line = ax.lines[0]
ax2.set_ylabel('Counts')
#ax2.grid(False)
        
        
fig.legend(labels=('Subscription rate','Counts'),loc='upper right',bbox_to_anchor=(1.1,1.1),
                   bbox_transform=ax.transAxes,fancybox=True, framealpha=0.2)
sns.despine(left=True,bottom=True,)
plt.grid(axis = 'y')
plt.show();
```

A customer's subscription rate was found to be similar whether being contacted 1 or 2 times for the current campaign. However, starting with the third contact, the subscription rate decreases. It is unsurprising, as it seems intuitive that if a customer hasn’t subscribed the first few times you’ve called them, they would be less likely to subscribe. 

Since one contact has a much larger market segment than two contacts, it shows that much fewer customers have received two communications. Thus, it seems wise to pursue customers at least for a second contact, if not beyond, and without proper follow-up beyond a single contact, there might be a loss in potential overall subscriptions. However, it’s important to note that many of those contacted only once may have given a firm answer, so analysis would need to be done to see what % of these customers who received a single call were simply not followed up on despite not giving a firm answer. 


##### Statistic Significance

__Hypothesis__:
  - H0 : Subscription and contact_number groups have no relationship (independent).
  - H1 : There is a relationship (dependent) between Subscription and contact_number groups.

```python
campaign_group_data = pd.crosstab(bank_cleaned.campaign_contact_number, bank_cleaned.subscription)
campaign_group_data
```

```python

chisq_and_posthoc_corrected(campaign_group_data)
```

The difference between the subscription rate for group with 2 contacts and the rest of the groups, except the group with only 1 contact, was found to be statistically significant with a p-value << .001.

#### Categorical Features Correlations With Subscription<a id='Categorical_Features_Correlations_With_Subscription'></a>

```python
bank_cat = bank_cleaned[[i for i in bank_cleaned.columns if i not in ('age','balance','month_num', 'pdays', 'day','previous', 'campaign')]]
bank_cat.head()

```

```python
label = preprocessing.LabelEncoder()

bank_encoded = pd.DataFrame() 

for i in bank_cat.columns :
  bank_encoded[i]=label.fit_transform(bank_cat[i])
```

```python
bank_encoded.head()
```

```python
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x,y)
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))

rows= []

for x in bank_encoded:
  col = []
  for y in bank_encoded:
    cramers =cramers_v(bank_encoded[x], bank_encoded[y]) 
    col.append(round(cramers,2))  
  rows.append(col)
  
cramers_results = np.array(rows)
df = pd.DataFrame(cramers_results, columns = bank_encoded.columns, index =bank_encoded.columns)
df
```

```python
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(15, 9))
    ax = sns.heatmap(df, annot=True,cmap = 'RdYlBu', linewidths=0.2,annot_kws={'size':10})
```

‘Previous campaign outcome’ strongly correlates with ‘pdays_group’ (how many days have passed since the last campaign). However, the majority of the data in the two features is missing. ‘education’ also strongly correlates with ‘job.’ ‘age_group’ also positively correlates with ‘job’ and ‘marital.’ 

Turning the attention to the target feature,’ subscription,’ we see quite a few reasonable correlations. ‘previous campaign outcome’ and ‘pdays_group’  stand out, even though the majority of the data is missing. ‘age_group,’ ‘contact_type,’ ‘job,’ and ‘housing loan’ are also positively correlated with subscriptions.
